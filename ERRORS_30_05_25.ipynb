```python
#MEASUREMENT EORROR
```

```python
import numpy as np
import matplotlib.pyplot as plt

true_temp = 25
observed_temps = np.array([24.7, 25.2, 25.1, 24.9, 25.3])
measurement_errors = observed_temps - true_temp

print("Measurement Errors:", measurement_errors)

plt.scatter(range(1, 6), measurement_errors)
plt.axhline(0, color='r', linestyle='--')
plt.xlabel('Measurement Number')
plt.ylabel('Measurement Error (Â°C)')
plt.title('Measurement Errors in Temperature Readings')
plt.show()
```

```python
#STANDARD ERROR
```

```python
import numpy as np

sample = [102, 98, 100, 105, 97, 99, 101, 103, 95, 100]
n = len(sample)
sample_std = np.std(sample, ddof=1)  # sample standard deviation

standard_error = sample_std / np.sqrt(n)
print(f"Sample standard deviation: {sample_std:.2f}")
print(f"Standard error of the mean: {standard_error:.2f}")
```

```python
# RESIDUAL ERROR
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Sample dataset: Square footage vs House price (in $1000s)
X = np.array([[1500], [1600], [1700], [1800], [1900], [2000], [2100]])
y_actual = np.array([300, 320, 340, 360, 375, 390, 410])  # Actual prices

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y_actual)

# Make predictions
y_pred = model.predict(X)

# Calculate residuals
residuals = y_actual - y_pred

# Print actual, predicted, and residuals
for i in range(len(X)):
    print(f"House {i+1}: Actual=${y_actual[i]*1000}, Predicted=${y_pred[i]*1000:.2f}, Residual=${residuals[i]*1000:.2f}")

# Plot residuals
plt.figure(figsize=(8, 5))
plt.scatter(X, residuals, color='red', label='Residuals')
plt.axhline(y=0, color='blue', linestyle='--')
plt.xlabel("Square Footage")
plt.ylabel("Residual Error (in $1000s)")
plt.title("Residual Plot: Actual - Predicted House Prices")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# MSE : MEAN SQUARE ERROR
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Sample data: Square footage vs House price (in $1000s)
X = np.array([[1500], [1600], [1700], [1800], [1900], [2000], [2100]])
y_actual = np.array([300, 320, 340, 360, 375, 390, 410])  # Actual prices

# Train the model
model = LinearRegression()
model.fit(X, y_actual)

# Make predictions
y_pred = model.predict(X)

# Calculate residuals
residuals = y_actual - y_pred

# Calculate Mean Squared Error
mse = mean_squared_error(y_actual, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

# Optionally print detailed info
for i in range(len(X)):
    print(f"House {i+1}: Actual=${y_actual[i]*1000}, Predicted=${y_pred[i]*1000:.2f}, Residual=${residuals[i]*1000:.2f}")

# Plot residuals
plt.figure(figsize=(8, 5))
plt.scatter(X, residuals, color='red', label='Residuals')
plt.axhline(y=0, color='blue', linestyle='--')
plt.xlabel("Square Footage")
plt.ylabel("Residual Error (in $1000s)")
plt.title("Residual Plot: Actual - Predicted House Prices")
plt.legend()
plt.grid(True)
plt.show()
```
