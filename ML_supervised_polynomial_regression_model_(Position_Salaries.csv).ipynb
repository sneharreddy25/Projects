SUPERVISED ML: POLYNOMIAL LINEAR REGRESSION 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("Position_Salaries.csv")
```

```python
x = data_set.iloc[:,1:2]
y = data_set.iloc[:,2]
```

```python
print(data_set.to_string())
```

```python
from sklearn.linear_model import LinearRegression
lin_regs = LinearRegression()
lin_regs.fit(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regs = PolynomialFeatures(degree = 4)
x_poly = poly_regs.fit_transform(x)
lin_regs_2 = LinearRegression()
lin_regs_2.fit(x_poly, y)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x, lin_regs.predict(x), color = "red")
plt.title("Bluff Detection Model(Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x, lin_regs_2.predict(poly_regs.fit_transform(x)), color = "red")
plt.title("Bluff Detection Model(Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
lin_pred = lin_regs.predict([[10.5]])
print(lin_pred)
```
